```python
""" Data Cleaning pour Airbnb_Open_Data """


import pandas as pd

df = pd.read_csv("../data/raw/Airbnb_Open_Data.csv")


```

```python
df.isnull().sum()
```

```python
""" On supprime la colonne Licence, car elle ne contient que des valeurs nulles (sauf 2) et ne nous sera pas utile, meme chose pour Last Review"""

df = df.drop('license', axis=1)
df = df.drop('last review', axis=1)
df = df.drop('number of reviews', axis=1)
df = df.drop('reviews per month', axis=1)
```

```python
""" On supprime les lignes pour lesquelles le prix, la latitude et la longitude ne sont pas renseign√©es."""

df.dropna(subset=['lat'], how = 'all', inplace=True)
```

```python
df.dropna(subset=['price'], how = 'all', inplace=True)
df.isnull().sum()
```
